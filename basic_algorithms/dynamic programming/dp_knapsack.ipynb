```python
import copy
def knapSack(W, weight, profit, n):
    A = [[0 for x in range(W + 1)] for x in range(n + 1)]
    Sol=[[ [] for x in range(W + 1)] for x in range(n + 1)]

    for i in range(n + 1):
        for w in range(W + 1):  #we look at the i th element,the index in weight and profit array is (i-1)
            if i == 0 or w == 0:
                A[i][w] = 0
            elif weight[i-1] <= w:
                A[i][w] = max(profit[i-1]+ A[i-1][w-weight[i-1]], A[i-1][w]) 
                if A[i-1][w]< profit[i-1]+ A[i-1][w-weight[i-1]]:
                    Sol[i][w]=copy.deepcopy(Sol[i-1][w-weight[i-1]])
                    Sol[i][w].append(i)
                    print(i,w,A[i][w],Sol[i][w],'循环2')
                else:
                    Sol[i][w]=Sol[i-1][w]
                    print(i,w,A[i][w],Sol[i][w],'循环2')
            else:
                A[i][w] = A[i-1][w]
                Sol[i][w]=copy.deepcopy(Sol[i-1][w])   #deepcopy in case it changes (in fact it will) the value of Sol[i-1][w]
                print(i,w,A[i][w],Sol[i][w],Sol[i-1][w],'循环3')


    return A[n][W], Sol[n][W]


# Driver code
val = [2, 5, 7, 8]
wt = [3, 6, 7, 8]
W = 10
n = len(val)
print(knapSack(W, wt, val, n))


# This code is contributed by Bhavya Jain
```
